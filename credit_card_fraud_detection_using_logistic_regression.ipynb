# machine learning 

```python
from sklearn.model_selection import train_test_split
```

```python
x = Default.drop('Class' ,axis = 1)
y=Default['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2 ,  random_state = 21 , stratify = y) # stratify = yes means distribute class (0/1) equally in test and train 
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# showing 0 and 1 are equally distributed in dependent train and test

print(y_train.value_counts(normalize = True).round(2))
print(' ')
print(y_test.value_counts(normalize = True).round(2))
```

```python
# if not equally distributed then use SMOTE for adding dummy values 

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=33 , sampling_strategy = 0.75)
x_res , y_res = sm.fit_sample(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
confusion_matrix(y_test,y_pred)

#left diagonal gives correctly classified values while right diagonal gives incorrect 
```
